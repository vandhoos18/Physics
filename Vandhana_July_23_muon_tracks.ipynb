```python
# fit a linear model to a set of muon tracks in CMS
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
muondata1=pd.read_csv('https://raw.githubusercontent.com/QuarkNet-HEP/coding-camp/main/data/muon_track_1.csv',skiprows=2)
muondata1.head(9)
```

```python
muondata1.plot(x='x', y='y', label="muontrack1")
plt.show()
```

```python
muondata2=pd.read_csv('https://raw.githubusercontent.com/QuarkNet-HEP/coding-camp/main/data/muon_track_2.csv',skiprows=2)
muondata2.head(9)
```

```python
muondata2.plot(x='x',y='y',color='magenta', label='muontrack2')
plt.show()
```

```python
muondata3=pd.read_csv('https://raw.githubusercontent.com/QuarkNet-HEP/coding-camp/main/data/muon_track_3.csv',skiprows=2)
muondata3.head(9)
muondata3.plot(x='x',y='y',color='purple', label='muontrack3')
plt.show()
```

```python
# FIRST MODEL USING POLYFIT
m,b = np.polyfit(muondata3['x'],muondata3['y'],1) # this helps to get a linear fit for the muondata3 graph.
```

```python
plt.plot(muondata3['x'],m*muondata3['x']+b) # plotting x and y, but here y=mx+b
plt.show()
```

```python
muondata4=pd.read_csv('https://raw.githubusercontent.com/QuarkNet-HEP/coding-camp/main/data/muon_track_4.csv',skiprows=2)
muondata4.head(9)
muondata4.plot(x='x',y='y',color='green', label='muontrack4')
plt.show()
```

```python
# Remember: numpy.polyfit(x, y, deg) is used to make a linear or quadratic fit for a given data.

a,b,c=np.polyfit(muondata4['x'],muondata4['y'],2) # this helps to get a quadratic fit for the muondata4 graph.
plt.plot(muondata4['x'],a*muondata4['x']**2+b*muondata4['x']+c) # plotting x and y, but here y=ax^2+bx+c
plt.show()
```

```python
m,b = np.polyfit(muondata4['x'],muondata4['y'],1)
plt.plot(muondata4['x'],m*muondata4['x']+b) # plotting muondata4 to make a line of best fit.
plt.show()
```

```python
# SECOND MODEL USING LINEAR REGRESSION

from sklearn.linear_model import LinearRegression
X = muondata3['x'].values.reshape((-1, 1))
Y = muondata3['y'].values.reshape((-1, 1))
model = LinearRegression().fit(X, Y)
plt.scatter(X, Y)
plt.plot(X, model.predict(X), color='red')
plt.show()
```
